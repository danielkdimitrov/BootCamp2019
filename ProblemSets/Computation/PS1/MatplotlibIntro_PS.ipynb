## Problem 1

```python
n = 100
def foolin_round(n):
    rndm = np.random.normal(0,1,(n,n))
    means = np.mean(rndm, axis = 0 )
    return np.var(means)
```

```python
N = np.arange(100,1100,100)
var = np.zeros(10)

for idx_n, n in enumerate(N):
    var[idx_n] = foolin_round(n)

plt.figure()
plt.plot(N,var)
plt.xlabel('N')
plt.ylabel('Variable')
```

## Problem 2

```python
domain = np.linspace(-2*pi, 2*pi,100) 

plt.subplot(131)
plt.plot(domain, np.sin(domain))
plt.title('sin(x)')

plt.subplot(132)
plt.plot(domain, np.cos(domain))
plt.title('cos(x)')

plt.subplot(133)
plt.plot(domain, np.arctan(domain))
plt.title('arctan(x)')
```

## Problem 4

```python
domain = np.linspace(0, 2*np.pi,50) 

[xmin, xmax, ymin, ymax] = [0, 2*pi, -2, 2]

plt.subplot(221)
plt.plot(domain, np.sin(domain), color ='g' )
plt.title('sin(x)')

plt.subplot(222)
plt.plot(domain, np.sin(2*domain),'--', color ='r')
plt.title('sin(2x)')

plt.subplot(223)
plt.plot(domain, 2*np.sin(domain),'--', color ='b')
plt.title('2sin(x)')

plt.subplot(224)
plt.plot(domain, 2*np.sin(2*domain),':', color ='m')
plt.title('2sin(2x)')

plt.axis([xmin, xmax, ymin, ymax])
plt.suptitle('Functions')
```
