```python
import numpy as np
from matplotlib import pyplot as plt
import scipy as scipy
import statistics as stat
from scipy.special import binom
```

```python
x =np.linspace(0,1,100)
count = 0

for n in range(4):
    for v in range(n+1):
        count = count + 1

        p = binom(n,v)*(x**v)*(1-x)**(n-v)
        plt.subplot(2,5,count)
        plt.plot(x,p,label="n = "+str(n)+", v= "+str(v))
        plt.title(r"$b({},{})(x)$".format(v,n))
        plt.axis([0, 1, 0, 1])
```

## Problem 5: The Rosenberg function


```python
def ros(x,y): 
    return (1 - x)**2 + 100*(y - x**2)**2

x = np.linspace(-1.5,1.5,200)
X,Y = np.meshgrid(x,x)
Z = ros(X,Y) 

ros(X, Y)

fig, ax = plt.subplots(1,2)
ax[0].pcolormesh(X, Y, Z, cmap="viridis") # Heat map.
ax[1].contour(X, Y, Z, 6, cmap="viridis",levels=14, linewidths=0.5) # Contour map.
```
