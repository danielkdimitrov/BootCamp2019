# DSGE Problem Set

## Problem 1

Plug in the guess policy function $K_{t+1} = A e^{z_t}K_t^\alpha$ into the corresponding Euler equation (49.) 

$$
\frac{1}{e^{z_t}K_t^\alpha - K_{t+1}} = \beta E_t \left( 
\frac{\alpha e^{z_{t+1}} K_{t+1}^{\alpha -1}}{e^{z_{t+1}}K_{t+1}^\alpha-K_{t+2}} \right)
$$

and you get 

$$
A = \beta \alpha
$$

## Problem 2

We have 

$$
u(c_t,l_t) = \ln c_t + a \ln(1-l_t)\\
F(K_t,L_t,z_t) = e^{z_t}K^{\alpha}_t L^{1-\alpha}_t 
$$

This implies the following system of equations (following the script eq (35.) to (41.))

$$
c_t = (1-\tau)(w_tl_t + (r_t -\delta)k_t) + k_t + T_t - k_{t+1} \\
\frac{1}{c_t} = \beta E_t \left\{ \frac{1}{c_{t+1}}((r_{t+1} - \delta)(1-\tau)+ 1) \right\}\\
\frac{a}{1-l_t} = \frac{w_t}{c_t}(1-\tau)\\
r_t = \alpha e^{z_t}k_t^{\alpha-1} l_t^{1-\alpha} \\
w_t = (1-\alpha) e^{zt}k_t^{\alpha} l_t^{-\alpha} \\
\tau(w_tl_t + (r_t -\delta)k_t) = T_t \\
z_t = (1-p_z)\bar{z} + p_zz_{t-1} + \epsilon^z_t; \quad 
\epsilon^z_t \sim \text{i.i.d}(0, \sigma^2_z)\\
$$

We cannot solve this model analythically. We have one more dimention of the decision problem (labor) and more variables that influence the optimal decision (interest rates, taxes, etc.). It is much more difficult to make a guess about the functional form of the value function.

## Problem 3

We have the functional forms

$$
u(c_t, l_t) = \frac{c_t^{1-\gamma}-1}{1 - \gamma} + a \ln(1 - l_t) \\
F_(K_t, L_t, z_t) = e^{z_t} K_t^{\alpha}L_t^{1-\alpha}
$$

This implies the following system of equations (following the script eq. (35.) to eq. (41.)): 

$$
c_t = (1 - \tau)[w_tl_t + (r_t -\delta)k_t] + k_t + T_t - k_{t+1} \\
c_t ^{-\gamma} = \beta E_t\left[ c_{t+1}^{-\gamma} [(r_{t+1} - \delta)(1 - \tau) + 1]\right] \\
\frac{a}{1-l_t} = c_t ^{-\gamma} w_t (1-\tau) \\
r_t  =  \alpha e^{z_t}  \left(\frac{l_t}{k_t} \right)^{1 - \alpha} \\
w_t = (1 - \alpha) e^{z_t} \left(\frac{k_t}{l_t} \right)^{\alpha} \\	
T_t = \tau[w_tl_t + (r_t - \delta)k_t] \\
z_t = (1-\rho_z)\bar{z} + \rho_z z_{t-1} + \epsilon^z_t; \quad \epsilon^z_t \sim i.i.d. (), \sigma^2_z)
$$

# Problem 4

Now we have 

$$
u(c_t, l_t) = \frac{c_t^{1-\gamma} - 1}{1- \gamma} + a \frac{(1 - l_t)^{1 - \xi}-1}{1 - \xi} \\
F(K_t, L_t, z_t) = e^{z_t}[\alpha K^{\eta}_t + (1 - \alpha)L^{\eta}_t]^{\frac{1}{\eta	}}
$$

This implies the following system of equations (following the script eq. (35.) to eq. (41.)): 

$$
	c_t = (1 - \tau)[w_tl_t + (r_t -\delta)k_t] + k_t + T_t - k_{t+1} \\
	c_t ^{-\gamma} = \beta E_t\left[ c_{t+1}^{-\gamma} [(r_{t+1} - \delta)(1 - \tau) + 1]\right] \\
	\frac{a}{(1-l_t)^{\xi}} = c_t ^{-\gamma} w_t (1-\tau) \\
	r_t = \alpha e^{z_t} k_t^{\eta - 1} [\alpha k_t^{\eta} + (1 - \alpha) l_t^{\eta}]^{\frac{1 - \eta}{\eta}} \\
	w_t = (1 - \alpha) e^{z_t} l_t^{\eta - 1} [\alpha k_t^{\eta} + (1 - \alpha) l_t^{\eta}]^{\frac{1 - \eta}{\eta}} \\
	 T_t = \tau[w_tl_t + (r_t - \delta)k_t] \\
	z_t = (1-\rho_z)\bar{z} + \rho_z z_{t-1} + \epsilon^z_t; \quad \epsilon^z_t \sim i.i.d. (), \sigma^2_z)
$$

# Problem 5

We have

$$
u(c_t) = \frac{c_t^{1-\gamma} - 1}{1- \gamma} \\
F(K_t, L_t, z_t) = K_t^{\alpha} (L_t e_t^{z_t})^{1 - \alpha}
$$

This implies the following system of equations (following the script eq. (35.) to eq. (41.), assuming $L_t = 1$, and dropping eq. (37.)): 

$$
c_t = (1 - \tau)[w_t + (r_t - \delta)k_t] + k_t + T_t - k_{t+1} \\
c_t ^{-\gamma} = \beta E_t \left[ c_{t+1}^{-\gamma} [(r_{t+1} - \delta)(1 - \tau) + 1] \right] \\
r_t = \alpha \left(\frac{e^{z_t}}{k_t} \right)^{1 - \alpha} \\
w_t = (1 - \alpha) k_t ^{\alpha} e_t^{z_t(1 - \alpha)} \\
T_t = \tau [w_t + (r_t - \delta)k_t ] \\
z_t = (1 - \rho_z)\bar{z} + \rho_z z_{t-1} + \epsilon^z_t; \quad \epsilon^z_t \sim i.i.d (0, \sigma^2_z) 
$$

In the steady state we have 

$$
c_t = c_{t+1} = \bar{c} \\
k_t = k_{t+1} = \bar{k} \\
\epsilon_t = 0 \implies z_t = z_{t+1} = 0 \\
r_t = r_{t+1} = \bar{r} \\
w_t = w_{t+1} = \bar{w} \\
T_t = T_{t+1} = \bar{T}
$$

Pluggin in these variables in the system above, we can reduce the system to two equations and two unknown 

$$
\bar{c} = (1 - \tau)\left({\bar{w} + (\bar{r}-\delta) \bar{k}})\right)+\bar{T} \\
1 = \beta \left( (\bar{r} -\delta)(1-\tau) + 1 \right)\\
\bar{r} = \alpha \bar{k}^{\alpha-1} \\
\bar{w} = (1-\alpha)\bar{k}^\alpha \\
\bar{T} = \tau (\bar{w} + (\bar{r} - \delta)\bar{k})
$$



```python
import numpy as np
from scipy.optimize import fsolve
# pip install tabulate
from tabulate import tabulate
import pandas as pd

from collections import namedtuple
```

```python
def steady_st(args):
    k,c, w, r, T  = args
    f1 = c - (1-tau)*(w + (r-delta)*k) - T 
    f2 = 1 - beta* ((r-delta)*(1-tau)+1) 
    f3 = r - alpha*k**(alpha-1) 
    f4 = w - (1-alpha)*k**alpha
    f5 = T - tau*(w + (r-delta)*k)
    return (f1,f2,f3,f4,f5)

gamma = 2.5
beta= .98
alpha= .40
delta = .10
z = 0 
tau = .05

k0 = 10
c0 = 10
w0 = 10
r0 = .1
T0 = 5
args_0 = (k0,c0, w0, r0, T0 )
(k,c, w, r, T) = fsolve(steady_st,args_0)

F_bar = k**alpha
inv = delta*k 

print('---------------- k =',round(k,2),'-------------------------' )
print('---------------- c =',round(c,2),'---------------------------' )
print('---------------- w =',round(w,2),'---------------------------' )
print('---------------- r =',round(r,2),'---------------------------' )
print('---------------- T =',round(T,2),'---------------------------' )
print('---------------- output =',round(F_bar,2),'------------------' )
print('---------------- investment =',round(inv,2),'----------------' )
```

# Problem 6

We have

$$
u(c_t) = \frac{c_t^{1-\gamma} - 1}{1- \gamma} + a \frac{(1-l)^{1-\xi}-1}{1-\xi} \\
F(K_t, L_t, z_t) = K_t^{\alpha} (L_t e_t^{z_t})^{1 - \alpha}
$$

This implies the following system of equations (following the script eq. (35.) to eq. (41.), assuming $L_t = 1$, and dropping eq. (37.)): 

$$
c_t = (1 - \tau)[w_t l_t + (r_t - \delta)k_t] + k_t + T_t - k_{t+1} \\
c_t ^{-\gamma} = \beta E_t \left[ c_{t+1}^{-\gamma} [(r_{t+1} - \delta)(1 - \tau) + 1] \right] \\
a (1-l_t)^{-\xi} = c_t^{-\gamma}w_t(1-\tau) \\
r_t = \alpha \left(\frac{e^{z_t}l_t}{k_t} \right)^{1 - \alpha} \\
w_t = (1 - \alpha) \frac{k_t}{l_t} ^{\alpha} e_t^{z_t(1 - \alpha)} \\
T_t = \tau [w_t l_t + (r_t - \delta)k_t ] \\
z_t = (1 - \rho_z)\bar{z} + \rho_z z_{t-1} + \epsilon^z_t; \quad \epsilon^z_t \sim i.i.d (0, \sigma^2_z) 
$$

In the steady state we have 

$$
c_t = c_{t+1} = \bar{c} \\
k_t = k_{t+1} = \bar{k} \\
l_t = l_{t+1} = \bar{l} \\
\epsilon_t = 0 \implies z_t = z_{t+1} = 0 \\
r_t = r_{t+1} = \bar{r} \\
w_t = w_{t+1} = \bar{w} \\
T_t = T_{t+1} = \bar{T}
$$

Pluggin in these variables in the system above, we can reduce the system to two equations and two unknown 

$$
\bar{c} = (1 - \tau)\left({\bar{w}\bar{l} + (\bar{r}-\delta) \bar{k}})\right)+\bar{T} \\
1 = \beta \left( (\bar{r} -\delta)(1-\tau) + 1 \right)\\
a(1-\bar{l})^{-\xi} = \bar{c}^{-\gamma}\bar{w}(1-\tau) \\
\bar{r} = \alpha \left(\frac{\bar{k}}{\bar{l}} \right)^{\alpha-1} \\
\bar{w} = (1-\alpha)\left(\frac{\bar{k}}{\bar{l}}\right)^\alpha \\
\bar{T} = \tau (\bar{w} \bar{l} + (\bar{r} - \delta)\bar{k})
$$


```python
def steady_st_kl(args):
    k,c, l, w, r, T  = args
    f1 = c - (1-tau)*(w*l + (r-delta)*k) - T 
    f2 = 1 - beta* ((r-delta)*(1-tau)+1) 
    f3 = a*(1-l)**(-xi) - c**(-gamma)*w*(1-tau)
    f4 = r - alpha*(k/l)**(alpha-1)*np.exp(z*(1-alpha)) 
    f5 = w - (1-alpha)*(k/l)**alpha*np.exp(z*(1-alpha)) 
    f6 = T - tau*(w*l + (r-delta)*k)
    return (f1,f2,f3,f4,f5,f6)

gamma = 2.5
beta= .98
xi = 1.5
alpha= .40
a = .5
delta = .10
z = 0 
tau = .05

l0 = .5
args_0 = (k0, c0, l0, w0, r0, T0 )
(k, c, l, w, r, T) = fsolve(steady_st_kl,args_0)

F_bar = k**alpha*l**(1-alpha)
inv = delta*k 

print('---------------- k =',round(k,2),'-------------------------' )
print('---------------- c =',round(c,2),'---------------------------' )
print('---------------- l =',round(l,2),'---------------------------' )
print('---------------- w =',round(w,2),'---------------------------' )
print('---------------- r =',round(r,2),'---------------------------' )
print('---------------- T =',round(T,2),'---------------------------' )
print('---------------- output =',round(F_bar,2),'------------------' )
print('---------------- investment =',round(inv,2),'----------------' )
```

# Problem 7 


```python
def steady_st_kl(args, params):
    k,c, l, w, r, T  = args
    alpha, beta, gamma, delta, xi, tau, a, z  = params
    f1 = c - (1-tau)*(w*l + (r-delta)*k) - T 
    f2 = 1 - beta* ((r-delta)*(1-tau)+1) 
    f3 = a*(1-l)**(-xi) - c**(-gamma)*w*(1-tau)
    f4 = r - alpha*(k/l)**(alpha-1)*np.exp(z*(1-alpha)) 
    f5 = w - (1-alpha)*(k/l)**alpha*np.exp(z*(1-alpha)) 
    f6 = T - tau*(w*l + (r-delta)*k)
    return (f1,f2,f3,f4,f5,f6)

h = 0.01

k0 = 1.0
c0 = 1.0
l0 = .5
w0 = .5
r0 = .5
T0 = .5

args_0 = (k0, c0, l0, w0, r0, T0 )
params_stst = [.40,.98, 2.5, .10, 1.5, 0.5, .5, 0]

steady_st_kl_new = lambda args:steady_st_kl(args,params_stst)

f = fsolve(steady_st_kl_new,args_0)
```

```python
params = np.zeros_like(params_stst)

df = []
for i, param_i  in enumerate(params):
    params = params_stst.copy()
    params[i] = params[i]+h
    steady_st_kl_new = lambda args:steady_st_kl(args,params)
    f_h =  fsolve(steady_st_kl_new,args_0)
    df.append((f - f_h)/h)

param_list = {'alpha', 'beta', 'gamma', 'delta', 'xi', 'tau', 'a', 'z' }
arg_list = ['k','c', 'l', 'w', 'r', 'T']

print('--------------------------- Numerical Derivatives --------------------')
print(' ')

table = pd.DataFrame( index = param_list,data=df)
table.columns = arg_list
print(table)

```
